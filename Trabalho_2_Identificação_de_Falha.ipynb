Este dataset  composto de 1951 sries temporais multivariadas adquiridas por sensores em um simulador de falhas de mquinas (MFS)Alinhamento-Balano-Vibrao (ABVT) da SpectraQuest. O 1951 compreende seis diferentes estados simulados: **funo normal**, falha de desequilbrio, falhas de desalinhamento horizontal e vertical e falhas de rolamento interno e externo.

# Dataset
Detalhes do processo de aquisio dos dados e os dados esto desiponvei [aqui](https://www02.smt.ufrj.br/~offshore/mfs/page_01.html).

### Aquisio:


+ Three Industrial IMI Sensors, Model 601A01 accelerometers on the radial, axial and tangencial directions:

  + Sensibility
    (20%) 100 mV per g (10.2 mV per m/s2);

  + Frequency range
    (3 dB) 16-600000 CPM (0.27-10.000 Hz);

  + Measurement range
    50 g (490 m/s2).

+ One IMI Sensors triaxial accelerometer, Model 604B31, returning data over the radial, axial and tangencial directions:

  + Sensibility
    (20%) 100 mV per g (10.2 mV per m/s2);

  + Frequency range
    (3 dB) 30-300000 CPM (0.5-5.000 Hz);

  + Measurement range
    50 g (490 m/s2)

+ Monarch Instrument MT-190 analog tachometer

+ Shure SM81 microphone with frequency range of 20-20.000 Hz

+ Two National Instruments NI 9234 4 channel analog acquisition modules, with sample rate of 51.2 kHz

### Composio dos dados

Os dados esto separados em pastas coletados em situaes que descrevem casos normais e de falha. As falhas esto descritas por diversos tipos. Os dados contm sinais dos seguinte tipos:
   
  + normal function
  + imbalance fault
  + horizontal misalignment
  + vertical misalignment
  + inner bearing faults
  + outer bearing faults

Cada sinal est armazenado em um arquivo **.csv**. Contendo diversas linhas representando os dados coletados ao longo do tempo e **oito** colunas, as quais representam as seguintes informaes:

+ coluna 1:
  sinal do tacmetro que permite estimar a frequncia de rotao;

+ colunas 2 a 4: acelermetro de rolamento suspenso (direo axial, radiale tangencial);

+ colunas 5 a 7: acelermetro de rolamento saliente (direo axial, radiale tangencial);

+ coluna 8: microfone.

## Exemplo de um sinal normal

```python
import numpy as np
import pandas as pd
```

```python
x = np.array(pd.read_csv('COLOQUE AQUI O ENDEREO DO ARQUIVO CSV QUE CONTEM OS DADOS'))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(23, 8))
plt.plot(x[0:,7], color ="red")
```

```python
fig, axs = plt.subplots(8, sharex=False, sharey=False,figsize=(15,15))
fig.suptitle('All Features of Normal State')
for i in range(0, x.shape[1]):
    axs[i].plot(x[:,i])
    axs[i].set_title('{} Column of Dataset'.format(i))
```

## FFT

```python
x_fft = np.fft.fft(x[0:,7])
plt.figure(figsize=(23, 8))
x_fft_real = np.abs(x_fft)
plt.plot(x_fft_real[0:int(len(x_fft_real)/2)], color ="blue")
```
